```python
# Chat with an intelligent assistant in your terminal
from openai import OpenAI

# Point to the local server
client = OpenAI(base_url="http://localhost:1234/v1", api_key="lm-studio")

            
import pandas as pd
import re

# Function to extract and format year
def format_year(date_str):
    # Search for patterns that capture full year '2023' or '2024'
    match = re.search(r'\b(2023|2024)\b', date_str)
    if match:
        return match.group(0) + ' News. '  # Directly return the matched year

    # Search for patterns that capture '23' or '24' not surrounded by other digits
    match = re.search(r'(?<!\d)(23|24)(?!\d)', date_str)
    if match:
        return '20' + match.group(0) + ' News. ' # Convert '23' to '2023', '24' to '2024'

    return None

df = pd.read_csv('Sudan_Incident_Log_May.csv', encoding='utf-8-sig')

df['formatted_year'] = df['Date'].apply(format_year)

df['Meta_data']=None
for i, row in df.iterrows(): 
    print("--------------" + str(row['formatted_year']))
    text = str(row['formatted_year']) + str(row['Incident Narrative'])
    #print(text)
    message = [
        {"role": "system", "content": """Identify the geographic locations and times where the conflict incident occurred in the format 'neighborhood, state, country, mm/dd/yyyy'. If there are multiple locations, list each location and date on a separate line without any explanation."""},
        {"role": "user", "content": text},
    ]
    # Call API to generate responses
    completion = client.chat.completions.create(
        model="model-identifier",
        messages=message,
        temperature=1,
        stream=True,
    )
    output =''
    for chunk in completion:
        output += str(chunk.choices[0].delta.content)
    print(str(i) + " "+ output)
    df.at[i,'Meta_data'] = output
    #if i%100==0 and i!=0:
        #index=i/10
    df.to_csv(f'Sudan_Incident_Log_May_zero_shot.csv', index=False, encoding='utf-8-sig')
```
